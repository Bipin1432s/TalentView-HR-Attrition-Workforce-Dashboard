```python
# Example: fill numeric nulls with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Example: fill categorical nulls with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
# Check dtypes again
df.dtypes
```

```python
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    df[col] = df[col].str.strip()           # remove spaces
    df[col] = df[col].str.replace("_", " ") # optional
```

```python
df.nunique().sort_values()
```

```python
df.nunique().sort_values()
```

```python
df.drop(columns=["EmployeeCount", "Over18", "StandardHours"], inplace=True)
```

```python
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
data = df.copy()
```

```python
data["Attrition"].value_counts()
```

```python
data["Attrition"].value_counts(normalize=True) * 100
```

```python
cat_cols = data.select_dtypes(include="object").columns

for col in cat_cols:
    print(f"\n--- {col.upper()} ---")
    print(data[col].value_counts())
```

```python
num_cols = data.select_dtypes(include=["int64", "float64"]).columns

data[num_cols].describe().T
```

```python
def attrition_summary(col):
    summary = (
        data.groupby(col)["Attrition"]
        .value_counts()
        .unstack()
        .fillna(0)
    )
    summary["Attrition_Rate_%"] = (
        summary["yes"] / summary.sum(axis=1) * 100
    )
    return summary.sort_values("Attrition_Rate_%", ascending=False)

attrition_summary("Department")
```

```python
for col in ["JobRole", "OverTime", "MaritalStatus", "JobSatisfaction", "BusinessTravel"]:
    print(f"\n--- ATTRITION BY {col.upper()} ---")
    print(attrition_summary(col))
```

```python
data_corr = data.copy()
data_corr["attrition_flag"] = data_corr["Attrition"].map({"yes": 1, "no": 0})

corr = data_corr.select_dtypes(include="number").corr()

corr["attrition_flag"].sort_values(ascending=False)
```

```python
data["age_group"] = pd.cut(
    data["Age"],
    bins=[18, 25, 35, 45, 60],
    labels=["18-25", "26-35", "36-45", "46-60"]
)

attrition_summary("age_group")
```
